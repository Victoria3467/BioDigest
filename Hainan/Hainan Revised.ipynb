```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
get_ipython().magic('matplotlib inline')
sns.set(style='white', context='notebook', palette='deep')
plt.rcParams[ 'figure.figsize' ] = 9 , 5
```

```python
xl = pd.ExcelFile("HainanData_kindofClean.xlsx")
```

```python
df = xl.parse("Clean")
df.shape
```

```python
nan_rows = df[df['day #2'].isnull()]
nan_rows
```

```python
df.drop(['Day', 'Year', 'Water/m3', 'Total electricity cons (kWh)', '50% NaOH/kg', 'FeCl2/kg', 'PAM/kg', 'Defoamer/kg', 'day #2'], axis = 1, inplace = True)
df
```

```python
Month_CNG = df[['Month', 'BioCNG Sold (m3)']].groupby(['Month']).mean()
Month_CNG
```

```python
LABELS = Month_CNG.index
Month = list(range(13))
CNG_month = Month_CNG['BioCNG Sold (m3)']
a = []
for i in CNG_month:
    a.append(i)

plt.figure(figsize = (10,10))
plt.bar(Month, a, align='center')
plt.xticks(Month, LABELS)
plt.show()
```
